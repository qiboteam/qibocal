```python
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
```

```python
def sin(t, omega, amp, offset, phase):
    return amp * np.sin(t*omega + phase) + offset
```

```python
scale = 20
x = np.linspace(0, scale, 30)
y = np.sin(x) + np.random.rand(x.size) * 0.5

par, cov = scipy.optimize.curve_fit(sin, x, y, [1, np.quantile(y, 0.8) - np.quantile(y, 0.2), np.median(y), 0], maxfev=int(1e6))
xp = np.linspace(0, scale * 1.5, 1000)
plt.plot(xp, sin(xp, *par))
plt.scatter(x, y)

print((par[0] - 1)*100, (par[1] - 1)*100)
```
